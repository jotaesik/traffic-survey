```python
jijace_list = [["강남구" , 11680],
["강동구" , 11740],
["강북구" , 11305],
["강서구" , 11500],
["관악구" , 11620],
["광진구" , 11215],
["구로구" , 11530],
["금천구" , 11545],
["노원구" , 11350],
["도봉구" , 11320],
["동대문구" , 11230],
["동작구" , 11590],
["마포구" , 11440],
["서대문구" , 11410],
["서초구" , 11650],
["성동구" , 11200],
["성북구" , 11290],
["송파구" , 11710],
["양천구" , 11470],
["영등포구" , 11560],
["용산구" , 11170],
["은평구" , 11380],
["종로구" , 11110],
["중구" , 11140],
["중랑구" , 11260]]
```

```python
from tqdm import tqdm
```

```python
df.rename(columns={"GUBUN_NM" : "MONTH"})
```

```python
bic_age = pd.DataFrame()
```

```python
from tqdm import tqdm
```

```python
for x in tqdm(range(len(jijace_list))):

    for y in range(len(year_list)):
        payload["sYear"]=year_list[y]
        payload["eYear"]=year_list[y]
        payload["jijace"]=jijace_list[x][1]             # [1] 을 해야 지자체의 고유 코드를 넣는 거임
        r = requests.post(url, data=payload)
        bs = eval(r.text)
        df = pd.DataFrame(bs)
        df['YEAR']=year_list[y]                         # 년도 컬럼 추가하기
        bic_age = pd.concat([bic_age, df], axis=0)
```

### 2018~2022년 월별로 따릉이 이용건수 추세를 보면서 연령대별 교통사고 자전거와 연관성 확인

```python
# db에 있는 거 가져온거임
CycleRentUse = pd.read_sql_query("select * from CycleRentUse", con=con)
```

```python
CycleRentUse
```

```python
# 대여일 같을 경우에 대여수 합쳐서 나열...
CRU = CycleRentUse.groupby(['Rent_Date'], as_index=False)[['Rent_Count']].sum()
```

```python
CRU
```

```python
pip install matplotlib
```

```python
pip install seaborn
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 따릉이 이용자수 년월별 그래프
plt.figure(figsize=(25,10))
plt.plot(CRU['Rent_Date'], CRU['Rent_Count'], color='steelblue')
plt.title('Cycle Rent Use')
plt.xlabel('Rent_Date')
plt.ylabel('Rent_Count')
plt.show()
```

```python
# db에서 자전거 사고수 월별로 분류한 거 불러오기
age_accident_cycle_month = pd.read_sql_query("select * from age_accident_cycle_month", con=con)
```

```python
age_accident_cycle_month
```

```python
# 년, 지자체, 월, 사고수만 뽑기
accident_counts = age_accident_cycle_month.loc[:, ['YEAR','JIJACE_NM','MONTH','ACC_CNT']]
```

```python
accident_counts
```

```python
# 위에서 년, 월 동일한 것 합치기
accident_counts2 = accident_counts.groupby(['YEAR','MONTH']).sum().reset_index()
```

```python
accident_counts2
```

```python
# 지자체 빼기
# 위에서 지자체별 그래프 뽑고 싶어서 지자체 포함 df 뽑았는데 ... 어쩌다보니 패스함.....
accident_counts2 = accident_counts2.drop('JIJACE_NM', axis=1)
```

```python
accident_counts2
```

```python
# 월을 x 축으로 하기엔 겹치는 게 있어서 그래프 제대로 안 나와서 CRU이랑 합친 것
age_bic = pd.concat([CRU,accident_counts2],axis=1)
```

```python
age_bic
```

```python
# 월별로 자전거 사고수 추이 보는 그래프
plt.figure(figsize=(25,10))
plt.plot(age_bic['Rent_Date'], age_bic['ACC_CNT'], color='steelblue')
plt.title('Accident_Count')
plt.xlabel('MONTH')
plt.ylabel('Accident_Count_Month')
plt.show()
```

```python
# 월별로 따릉이 대여수 추이 보는 그래프
plt.figure(figsize=(25,10))
plt.plot(CRU['Rent_Date'], CRU['Rent_Count'], color='steelblue')

# 월별로 자전거 사고수 추이 보는 그래프
plt.figure(figsize=(25,10))
plt.plot(age_bic['Rent_Date'], age_bic['ACC_CNT'], color='steelblue')
```

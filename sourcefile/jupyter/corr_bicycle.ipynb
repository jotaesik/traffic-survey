```python
weather_acc_df['Yr_Mth'] = weather_acc_df['YEAR'].astype(str) + '-' + weather_acc_df['MONTH'].astype(str).str.zfill(2)
weather_acc_df['Yr_Mth'] = pd.to_datetime(weather_acc_df['Yr_Mth'], format='%Y-%m')
```

```python
acc_df = weather_acc_df.groupby(['Yr_Mth', 'JIJACE_NM'])['ACC_CNT'].sum().reset_index()
```

```python
hire_point_df.columns
```

```python
rent_df = pd.merge(hire_point_df, cycle_rent_use_df, how='left', left_on=['Unnamed: 0'], right_on=['Rent_Number'])
```

```python
rent_df.isnull().sum()
```

```python
rent_df.dropna(inplace=True)
```

```python
rent_df = rent_df.groupby(['Rent_Date', 'STA_LOC'])['Rent_Count'].sum().reset_index()
```

```python
rent_df
```

```python
rent_df['Rent_Date'] = pd.to_datetime(rent_df['Rent_Date'])
```

```python
corr_df = pd.merge(rent_df, acc_df, how='outer', left_on=['Rent_Date','STA_LOC'], right_on=['Yr_Mth', 'JIJACE_NM'])
```

```python
corr_df = corr_df.fillna(0)
```

```python
# Compute the correlation matrix
corr = corr_df[['Rent_Count', 'ACC_CNT']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".4f")
plt.title('Correlation Heatmap between Rent Count and Accident Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

gugun_list = corr_df['STA_LOC'].unique()

# 5x5 서브플롯에 각 구별로 상관관계 히트맵 그리기
fig, axes = plt.subplots(5, 5, figsize=(20, 20))

for i, gugun in enumerate(gugun_list):
    row = i // 5
    col = i % 5
    
    gugun_cor_df = corr_df[corr_df['STA_LOC'] == gugun]
    corr = gugun_cor_df[['Rent_Count', 'ACC_CNT']].corr()
    
    sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f", ax=axes[row, col], vmin=0, vmax=1)
    axes[row, col].set_title(f'Correlation Heatmap for {gugun}')

plt.tight_layout()
plt.show()
```
